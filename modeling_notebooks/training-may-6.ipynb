### Training Loop

```python
def model_train(train_dataloader, val_dataloader, model, criterion, optimizer, epochs=10): 

    for epoch in range(epochs): 
        train_loss = 0
        train_size = 0

        val_loss = 0
        val_size = 0
        # Training
        
        for step, (x1, x2, y) in enumerate(train_dataloader): 
            
            model.train()
            
            x1 = x1.to(device)
            x2 = x2.to(device)

            train_size += y.shape[0]

            optimizer.zero_grad()
            z, labels = model(x1, x2)
            loss = criterion(z, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()*y.shape[0]

        # Validation 
        for x1, x2, y in val_dataloader:
            
            model.eval()
            x1 = x1.to(device)
            x2 = x2.to(device)

            val_size += y.shape[0]

            z, labels = model(x1, x2)
            loss = criterion(z, labels)
            val_loss += loss.item()*y.shape[0]

        print(f"Epoch: {epoch+1}/{epochs} \t T Loss: {train_loss/train_size}\
          \t V Loss: {val_loss/val_size}")
```

```python
batch_size = 10
train_data = ContrastiveCIFAR10()
val_data = ContrastiveCIFAR10(mode = 'val')
```

```python
train_dl = DataLoader(train_data, batch_size)
val_dl = DataLoader(val_data, batch_size)
```

### Include Crops in the DataSet [5/8]

```python
# ~48 mins
model2 = SimCLR().to(device)
criterion = NTXentLoss(temperature=0.1).to(device)
optimizer = torch.optim.Adam(model2.parameters())

model_train(train_dl, val_dl, model2, criterion, optimizer)
```

```python
model1 = SimCLR().to(device)
criterion = NTXentLoss(temperature=0.1).cuda()
optimizer = torch.optim.Adam(model1.parameters())

model_train(train_dl, val_dl, model1, criterion, optimizer)
```
